## **Converting the weather categories into standard categories**

```python
x = 'Thunderstorms, Moderate Rain Showers, Fog'

lists_of_lists = [w.split() for w in x.split(',')]
lists_of_lists
```

```python
from itertools import chain
flat_list = list(chain(*lists_of_lists))
flat_list
```

```python
def Create_List(x):
  lists_of_lists = [w.split() for w in x.split(',')]
  flat_list = list(chain(*lists_of_lists))
  return flat_list
```

```python
def Get_Weather(list1):
    if 'Fog' in list1 and 'Rain' in list1:
      return 'RAIN+FOG'
    elif 'Snow' in list1 and 'Rain' in list1:
      return 'RAIN+SNOW'
    elif 'Snow' in list1:
      return 'SNOW'
    elif 'Rain' in list1:
      return 'Rain'
    elif 'Fog' in list1:
      return 'Fog'
    elif 'Clear' in list1:
      return 'Clear'
    elif 'Cloudy' in list1:
      return 'Cloudy'
    else:
      return 'RAIN'
```

```python
Create_List(x)
```

```python
Get_Weather(Create_List(x))
```

```python
data['Std_Weather'] = data['Weather'].apply(lambda x : Get_Weather(Create_List(x)))
data.head()
```

```python
data.Std_Weather.value_counts()
```
